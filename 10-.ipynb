# What now?
Now that you have learned the basics of Python, you may be wondering what to do next. There are many different directions you can take your learning, depending on your interests and goals.
A small list of good resources for different fields can be found here:
https://github.com/SkalskiP/courses?tab=readme-ov-file
 Here are a few suggestions for what to do next:

## 0. Practice, practice, practice
The best way to solidify your understanding of Python is to practice writing code. You can freely use this set of notebooks to practice and experiment with the concepts you have learned. Feel free to distribute it with your friends and colleagues. If you want to use this to teach others, feel free to fork it on github. You can also find many coding challenges online that can help you practice your skills and learn new concepts. Some popular websites for coding challenges include LeetCode/neetcode(free), HackerRank, and Codewars. 

## 1. Algorithms and data structures
Algorithms and data structures are fundamental concepts in computer science. They are used to solve problems and to organize data in a way that allows for efficient access and manipulation. There are many resources online teaching you the most important concepts of this. Learning about this is very important if you are interested in big-data manipulation and efficient computing.

### Resources
1. https://github.com/tayllan/awesome-algorithms: 
    This github repository contains a curated list of algorithms and data structures resources, including books, tutorials, and implementations in various programming languages. It is a great resource for learning about algorithms and data structures and for finding implementations in different programming languages.
2. https://runestone.academy/ns/books/published/pythonds3/index.html: 
    This is a great resource for learning about algorithms and data structures. It is a free online book that covers a wide range of topics, including sorting algorithms, searching algorithms, and data structures such as lists, stacks, queues, and trees. It is also supported by youtube videos:
    https://www.youtube.com/user/gjenkinslbcc


## 2. Machine learning
Machine learning is a subfield of artificial intelligence that focuses on the development of algorithms that can learn from and make predictions on data. It is a powerful tool for solving complex problems in various domains, including computer vision, natural language processing, and bioinformatics. If you are interested in machine learning, there are many resources available online to help you get started.

### Resources
1. https://developers.google.com/machine-learning/crash-course
This is a free online course that provides a comprehensive introduction developed by google to machine learning. It covers the fundamentals of machine learning, including important base concepts like loss, gradient descent and regression. It goes deeper into Neural networks and Large language models as well. The course is designed for beginners and includes interactive exercises and projects to help you apply what you've learned.

## 3. Computer Vision
